```python
import csv

if __name__ == '__main__':
    f = open('subway.csv')
    data = csv.reader(f)
    header = next(data)

    sub1 = {}
    sub2 = {}
    sub3 = {}
    sub4 = {}

    for row in data:
        if row[1] == '1호선':
            station = row[3]
            if station in sub1:
                total = int(row[4].replace(',', '')) + int(row[5].replace(',', ''))
                sub1[station] += total
            else:
                total = int(row[4].replace(',', '')) + int(row[5].replace(',', ''))
                sub1[station] = total

        if row[1] == '2호선':
            station = row[3]
            if station in sub2:
                total = int(row[4].replace(',', '')) + int(row[5].replace(',', ''))
                sub2[station] += total
            else:
                total = int(row[4].replace(',', '')) + int(row[5].replace(',', ''))
                sub2[station] = total

        if row[1] == '3호선':
            station = row[3]
            if station in sub3:
                total = int(row[4].replace(',', '')) + int(row[5].replace(',', ''))
                sub3[station] += total
            else:
                total = int(row[4].replace(',', '')) + int(row[5].replace(',', ''))
                sub3[station] = total

        if row[1] == '4호선':
            station = row[3]
            if station in sub4:
                total = int(row[4].replace(',', '')) + int(row[5].replace(',', ''))
                sub4[station] += total
            else:
                total = int(row[4].replace(',', '')) + int(row[5].replace(',', ''))
                sub4[station] = total

    sub1 = sorted(sub1.items(), 
                                  reverse=True, 
                                  key=lambda item: item[1])
    sub2 = sorted(sub2.items(), 
                                  reverse=True, 
                                  key=lambda item: item[1])
    sub3 = sorted(sub3.items(), 
                                  reverse=True, 
                                  key=lambda item: item[1])
    sub4 = sorted(sub4.items(), 
                                  reverse=True, 
                                  key=lambda item: item[1])

    print("*** Subway Report for Seoul on March 2023 ***")
    print("Line 1:")
    print(f"Busiest Station: {sub1[0][0]} ({sub1[0][1]})")
    print(f"Least used Station : {sub1[-1][0]} ({sub1[-1][1]})")

    print()
    print("Line 2:")
    print(f"Busiest Station: {sub2[0][0]} ({sub2[0][1]})")
    print(f"Least used Station : {sub2[-1][0]} ({sub2[-1][1]})")

    print()
    print("Line 3:")
    print(f"Busiest Station: {sub3[0][0]} ({sub3[0][1]})")
    print(f"Least used Station : {sub3[-1][0]} ({sub3[-1][1]})")

    print()
    print("Line 4:")
    print(f"Busiest Station: {sub4[0][0]} ({sub4[0][1]})")
    print(f"Least used Station : {sub4[-1][0]} ({sub4[-1][1]})")
```
